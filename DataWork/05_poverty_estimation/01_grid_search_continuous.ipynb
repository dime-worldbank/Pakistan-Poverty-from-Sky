## Load/Prep Data and Run Models

```python
df = pd.read_csv(s3.open('{}/{}'.format(bucket, os.path.join('OPM', 'FinalData', 'Merged Datasets', 'cnn_merge.csv'))))
```

```python
df = df[df['pscores'].notna()]
```

```python
import time
start = time.time()
print("started")

count = 1
for target in ['pscores']:

    print(target) # Print Status

    x = df.filter(regex='^cnn_', axis=1)
    y = df[target]

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=TEST_SIZE)

    # Normalize
    x_scaler = StandardScaler().fit(x_train)

    x_train = x_scaler.transform(x_train)
    x_test = x_scaler.transform(x_test)

    # Train/Evaluate -------------------------------------------
    parameters = grids.GRID_REGRESS

    # r_df: dataframe of results. Contains "model_number" variable to match with pred_df
    # pred_df: predicted values (also contains true value). In format of y_pred_[model_number] 
    r_df, pred_df = train_models(parameters, x_train, x_test, y_train, y_test, verbose=False)

    r_df['target'] = target
    r_df.to_csv(os.path.join(LOCAL_DIR, 'results_' + str(count) + '.csv'))
    boto3.Session().resource('s3').Bucket(bucket).Object(os.path.join('Laniah','Poverty Estimation Results', 'continuous', 'individual_files', 'results_1' + '.csv')).upload_file(os.path.join(LOCAL_DIR, 'results_' + str(count) + '.csv'))
    
    pred_df['target'] = target
    pred_df.to_csv(os.path.join(LOCAL_DIR, 'results_' + str(count) + '.csv'))
    boto3.Session().resource('s3').Bucket(bucket).Object(os.path.join('Laniah','Poverty Estimation Results', 'continuous', 'individual_files', 'results_2' + '.csv')).upload_file(os.path.join(LOCAL_DIR, 'results_' + str(count) + '.csv'))
    
    count = count + 1

end = time.time()
print('ended')
print("The time elapsed from start to end was", end - start)
```
